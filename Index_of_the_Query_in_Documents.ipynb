```python
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "The lazy dog sleeps.",
    "A quick brown dog outpaces a quick fox."
]
import re
def build_comprehensive_positional_index(documents):
    positional_index = {}

    for doc_id, text in enumerate(documents):
        clean_text = re.sub(r'[^\w\s]', '', text.lower())
        words = clean_text.split()

        for position, term in enumerate(words):
            if term not in positional_index:
                positional_index[term] = {}

            if doc_id not in positional_index[term]:
                positional_index[term][doc_id] = []

            positional_index[term][doc_id].append(position)

    return positional_index


index = build_comprehensive_positional_index(documents)
print(index)

# import json
# print(json.dumps(index, indent=4))
```

```python
def get_index_for_specific_query(query,full_index):
    terms = query.lower().split()

    for term in terms:
        if term not in full_index:
            return f"Term '{term}' not found in any document."

    results = {}

    common_docs = set(full_index[terms[0]].keys())
    for term in terms[1:]:
        common_docs &= set(full_index[term].keys())

    for doc_id in common_docs:
        match_positions = []
        pos_lists = [full_index[term][doc_id] for term in terms]

        for start_pos in pos_lists[0]:
            is_phrase = True
            for i in range(1, len(terms)):
                if (start_pos + i) not in pos_lists[i]:
                    is_phrase = False
                    break
            if is_phrase:
                match_positions.append(start_pos)

        if match_positions:
            results[doc_id] = match_positions

    return results

# query="quick brown"
query="lazy dog"
specific_results=get_index_for_specific_query(query,index)

print(f"Index for query '{query}':")
for doc_id,positions in specific_results.items():
    print(f"Document {doc_id}: Starts at position(s) {positions}")
```
